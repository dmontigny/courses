```python
import plotly.graph_objs as go 
from plotly.offline import init_notebook_mode,iplot
init_notebook_mode(connected=True) 
import pandas as pd
df = pd.read_csv('2014_World_GDP')
df.head()
```

```python
data = dict(type = 'choropleth',
            locations = df['CODE'],
            locationmode = 'ISO-3',
            colorscale= 'Portland',
            text= df['COUNTRY'],
            z=df["GDP (BILLIONS)"],
            colorbar = {'title':'World GDP'})
layout = dict(geo = {'scope':'world'})
choromap = go.Figure(data = [data],layout = layout)
iplot(choromap, validate=False)
```

```python
df2 = pd.read_csv('2014_World_Power_Consumption')
df2.head()
```

```python
data2 = dict(type = 'choropleth',
            locations = df2['Country'],
            locationmode = 'country names',
            colorscale= 'Portland',
            text= df2['Text'],
            z=df2["Power Consumption KWH"],
            colorbar = {'title':'World Power Use'})
layout2 = dict(geo = dict(showframe=False))
choromap = go.Figure(data = [data2],layout = layout2)
iplot(choromap, validate=False)
```

```python
df3 = pd.read_csv('2012_Election_Data')
df3.head()
```

```python
data3 = dict(type='choropleth',
            colorscale = 'YlOrRd',
            locations = df3['State Abv'],
            z = df3['Voting-Age Population (VAP)'],
            locationmode = 'USA-states',
            text = df3['State'],
            marker = dict(line = dict(color = 'rgb(255,255,255)',width = 2)),
            colorbar = {'title':"2012 Elections"}
            ) 
layout3 = dict(title = '2012_Election_Data',
              geo = dict(scope='usa',
                         showlakes = True,
                         lakecolor = 'rgb(85,173,240)')
             )
choromap = go.Figure(data = [data3],layout = layout3)
iplot(choromap)
```
