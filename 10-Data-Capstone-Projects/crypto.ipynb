```python
from pandas_datareader import data, wb
import pandas_datareader as pdr
import pandas as pd
import numpy as np
import datetime
import seaborn as sns
import matplotlib as plt
%matplotlib inline
```

```python
start = datetime.datetime(2017, 10, 1)
end = datetime.datetime(2019, 5, 6)
tickers = ['BTC-USD','ETH-USD','LTC-USD']
tickers
```

```python
BTC = pdr.DataReader('BTC-USD', 'yahoo', start, end)
BTC.head(5)
```

```python
ETH = pdr.DataReader('ETH-USD', 'yahoo', start, end)
ETH.head(5)
```

```python
LTC = pdr.DataReader('LTC-USD', 'yahoo', start, end)
LTC.head(5)
```

```python
lBTC = len(BTC.index)
lETH = len(ETH.index)
lLTC = len(LTC.index)
```

```python
print (lBTC, lETH, lLTC)
```

```python
ETH.info()
```

```python
bank_stocks = pd.concat([BTC, ETH, LTC], keys=tickers, axis=1, ignore_index=True)
bank_stocks.head()
```

```python
bank_stocks.columns.names = ['Bank Ticker','Stock Info']
```
