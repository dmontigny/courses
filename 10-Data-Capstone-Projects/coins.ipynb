```python
from pandas_datareader import data, wb
import pandas_datareader as pdr
import pandas as pd
import numpy as np
import datetime
import seaborn as sns
import matplotlib as plt
%matplotlib inline
```

```python
start = datetime.datetime(2017, 1, 1)
end = datetime.datetime(2019, 5, 6)
```

```python
tickers = ['BTC-USD','ETH-USD','LTC-USD']
tickers
```

```python
BTC = pdr.DataReader('BTC-USD', 'yahoo', start, end)
BTC.head(5)
```

```python
BTC.info()
```

```python
ETH = pdr.DataReader('ETH-USD', 'yahoo', start, end)
ETH.head(5)
```

```python
ETH.info()
```

```python
LTC = pdr.DataReader('LTC-USD', 'yahoo', start, end)
LTC.head(5)
```

```python
LTC.info()
```

Group all ticker info into single dataframe

```python
bank_stocks = pd.concat([BTC, ETH, LTC], keys=tickers, axis=1)
bank_stocks.head(5)
```

```python
bank_stocks.columns.names = ['Bank Ticker','Stock Info']
bank_stocks.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline

# Optional Plotly Method Imports
import plotly
import cufflinks as cf
cf.go_offline()
```

```python
for tick in tickers:
    bank_stocks[tick]['Close'].plot(figsize=(12,4),label=tick)
plt.legend()
```
